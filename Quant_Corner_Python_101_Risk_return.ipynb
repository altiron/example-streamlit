Now, let's develop the code to compute the skewness of a series of numbers.

Recall that the skewness is given by:

$$ S(R) = \frac{E[ (R-E(R))^3 ]}{\sigma_R^3} $$

```python
def skewness(r):
    demeaned_r = r - r.mean()
    # use the population standard deviation, so set dof=0
    sigma_r = r.std(ddof=0)
    exp = (demeaned_r**3).mean()
    return exp/sigma_r**3
```

```python
skewness(prices[['Pct']])
```

##  Kurtosis


$$ K(R) = \frac{E[ (R-E(R))^4 ]}{\sigma_R^4} $$

```python
def kurtosis(r):
    demeaned_r = r - r.mean()
    # use the population standard deviation, so set dof=0
    sigma_r = r.std(ddof=0)
    exp = (demeaned_r**4).mean()
    return exp/sigma_r**4
```

```python
kurtosis(prices[['Pct']]).sort_values()
```

```python
scipy.stats.kurtosis(prices['Pct'])
```

```python
def semideviation(returns):
    """
    Returns the semideviation aka negative semideviation of r
    r must be a Series or a DataFrame, else raises a TypeError
    """
    returns_negative = returns < 0
    return returns[returns_negative].std(ddof=0)
```

# Max Drawdown

```python
def drawdown(series: pd.Series):
    """Takes a time series of asset price.
       returns a DataFrame with columns for
       the previous peaks, and
       the percentage drawdown
    """
    previous_peaks = series.cummax()
    drawdowns = (series - previous_peaks)/previous_peaks
    return pd.DataFrame({"series": series,
                         "Previous Peak": previous_peaks,
                         "Drawdown": drawdowns})
```

```python
drawdown(returns.value)
```
